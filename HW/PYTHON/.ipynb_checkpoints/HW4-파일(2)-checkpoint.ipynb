```python
# 과제실습5 : 파일 - 실습2

import os

file_name = input("Enter a file name: ")

if os.path.exists(file_name):
    with open(file_name, "r") as fr:
        for line in fr:
            print(line.upper(), end="")
else:
    print('File not found')
```
