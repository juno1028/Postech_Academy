## Weibull 분포

```python
# weibull 분포에서 확률 계산하기

x = 2500
alpha = 5.2
beta = 2200

prob = stats.weibull_min.cdf(x, alpha, scale = beta)
print("P(X >= x): {:.3f}".format((1-prob)))
```

: 신뢰성 분야에서 자주 사용되는 분포가 바로 Weibull 분포이다. 2500시간 이상 작동할 확률을 구하는 것이므로, prob 값은 2500시간 이하로 작동할 확률이므로, 1에서 prob 값을 뺀 0.143이 알맞은 작동확률이다.

## 이항분포

```python
# 이항 분포

n = 5
for i in range(n+1):
    # 이항분포 Probability Mass Function
    prob = stats.binom.pmf(k=i, n=n, p=0.3)
    print("P(X={0}) = {1:.3f}".format(i,prob))
```

: n이 무한대로 가면, 이항분포는 정규분포에 가까워진다. 이때, 광택도 불량 확률이 30%(0.3)이므로 5대의 차량을 임의로 선택했을 때는 이항분포를 따른다.

## 포아송 분포

```python
# 평균
m = 3
# 1) 2명의 손님이 도착할 확률
prob = stats.poisson.pmf(2,m)
print("{0}명의 손님이 도착할 확률 : {1:.1f}".format(2, prob*100))

# 2) 1명 이하의 손님이 도착할 확률
cdf_prob = stats.poisson.cdf(1,m)
print("{0}명 이하의 손님이 도착할 확률: {1:.1f}".format(1, cdf_prob*100))

# 3) 3명 이상의 손님이 도착할 확률
cdf_prob_2 = 1 - stats.poisson.cdf(3,m)
print("{0}명 이상의 손님이 도착할 확률: {1:.1f}".format(3, cdf_prob_2*100))
```

: 일정한 시간 또는 일정한 공간에서 발생하는 성공횟수에 대한 이산확률분포가 포아송 분포이다. 10시에서 11시 사이라는 일정한 시간동안 발생하는 횟수를 세는 것이므로, 포아송 분포가 적절하다. 2명의 손님이 도착할 확률은 특정 사실이 2번 발생하는 것이므로, pmf를 사용하고, 1명 이하의 손님은 cdf로 1, 3명 이상의 손님은 cdf에 3을 넣은 값을 1에서 뺀 값을 넣었다.

## 정규성 검정

```python
# 데이터 불러오기
ds_mat = pd.read_csv("/home/piai/test/Practice/data/2sample_t_test.csv", engine="python")
ds_mat.head()
# 슬라이싱해서 데이터가 있는 부분만 따로 가져온다.
data1 = ds_mat['Mat_A'][0:10]
data2 = ds_mat['Mat_B'][0:10]
#정규성 검정 - Shapiro-Wilk 테스트: Mat_A 에 대해서
statistic1, p1 = stats.shapiro(data1)
print("Mat_A_Shapiro-Wilk Tst: statistic={}, p-value={}".format(statistic1, p1))
#정규성 검정 - Shapiro-Wilk 테스트: Mat_B 에 대해서
statistic2, p2 = stats.shapiro(data2)
print("Mat_B_Shapiro-Wilk Tst: statistic={}, p-value={}".format(statistic2, p2))
```

: 회귀 분석 등에서는 확률분포가 정규분포를 따르는지 아닌지를 확인하는 것이 주용하다. 이러한 검정을 정규성 검정(normality test)이라고 한다. p-value가 0.05보다 크면 정규로 판단하는데, Mat_A와 Mat_B 모두 정규성을 검정해보았을 때, p-value가 0.05보다 크므로 정규성을 가진다고 판단할 수 있다.

## 모평균 추정 (모표준편차 아는 경우)

```python
df = pd.DataFrame({"sample":[18,18,20,21,20,23,19,18,17,21,22,20,20,20,21,20,19,19,18,17,19]})
df.head()
```

```python
lower, upper = stats.norm.interval(0.95, loc=np.mean(df), scale=3.8/np.sqrt(20))
print("신뢰구간: ({0},{1})".format(lower.round(2), upper.round(2)))
```

: 모평균에 신뢰도 구간을 구할 때, interval을 사용하여 구한다. 95% 신뢰구간이므로, 0.95를 parameter로 넣고, scale은 모표준편차를 표본 갯수의 제곱근으로 나눈값을 대입하여 구했다. 답을 해석하자면, 표본을 랜덤추출 하였을 때, 17.86과 21.19 사이에 있을 확률이 95%라는 뜻이다. 

## 모평균 추정(모표준편차 모르는 경우)

```python
df = pd.DataFrame({"sample": [73,71,74,69,70,73,70,68,75,72,70,72,73,70,70,72,71,70,75,72]})

lower, upper = stats.t.interval(0.95, len(df)-1, loc = np.mean(df), scale = scipy.stats.sem(df))
print("신뢰구간: ({0},{1})".format(lower.round(2), upper.round(2)))
```

: 대부분의 경우, 모 표준편차를 모른다. 따라서 표본으로부터 표준편차를 추정해야만 한다. t분포의 신뢰구간을 구하였고, 신뢰구간 0.95, 모표준편차를 모르므로, len(df)-1, 표본표분편차를 활용하여 값을 구했다. 그 결과, 70.6, 72.6의 신뢰구간을 얻었으며, 이는 표본을 샘플링하였을 때, 70.6과 72.6사이에 있을 확률이 95%라는 것이다.

## 평균 검정(모집단 표준편차를 모르는 경우)

```python
df = pd.DataFrame({"sample": [85.0, 79.0, 79.1, 79.9, 81.6, 78.6, 85.4, 83.4, 78.1, 79.2]})
t_result = stats.ttest_1samp(df,78.0)

t,p=t_result.statistic.round(3), t_result.pvalue.round(3)

print("1-Sample t-test")
print("t통계량 : {}".format(t))
print("p-value : {}".format(p))
```

: 평균 검정의 절차는 다음과 같다. 첫째, 가설 수립 및 유의수준 설정, 둘째, 가설 검정 수행, 셋째, 검정 결과 판단이다. 이때, p-value가 0.05보다 작으면 변화가 없다고 판단한다. 위 문제에서는 p-value가 0.05보다 작으므로, 개선활동이 만족도를 변화시켰다고 보기 힘들다.

## 2 Sample t-test

```python
# 2 Sample t-test
df = pd.DataFrame({"A":[1.883,1.715,1.799,1.768,1.711,1.832,1.427,1.344],"B":[1.435,1.572,1.486,1.511,1.457,1.548,1.404,1.883]})

df.boxplot(['A', 'B'])
```

```python
t_result = stats.ttest_ind(df['A'], df['B'], equal_var = True)

# 2-Sample t-test의 t값 p값 저장
t, p = t_result.statistic.round(3), t_result.pvalue.round(3)

# 2-Sample t-test의 t값 p값 출력
print("2-Sample t-test")
print(" t검정통계량: {}".format(t))
print(" p-value : {}".format(p))
```

: 결론 : p값이 0.111로 유의수준 5%이상이므로 공급자 간에 모평균에 차이가 없다.

## Paired t-test

```python
df1 = pd.DataFrame({"before":[720,589,780,648,720,589,780,648,780,648]})
df2 = pd.DataFrame({"after":[710,580,787,712,750,600,782,670,790,680]})

# Paired t-test
t_result = stats.ttest_rel(df1, df2)

# Paired t-test의 t값 p값 저장
t, p = t_result.statistic.round(3), t_result.pvalue.round(5)

# Paired t-test의 t값 p값 출력
print("Paired t-test")
print(" t검정통계량: {}".format(t))
print(" p-value : {}".format(p))
```

: Paired t-test를 통해 짝을 이루는 집단의 평균의 차이가 있는지 검정하였다. 그 결과, p-value가 0.05보다 작았고, 따라서 점수 차이가 있다고 할 수 있다.

## Proportion test

```python
from statsmodels.stats.proportion import proportions_ztest

# 가설 설정
# H0 : p = 0.5
# H1 : p != 0.5

#Proportion test
count = 40
n_obs = 100
value = 0.5

# Proportion test 실행
stat,pval = proportions_ztest(count, n_obs, value)

# Proportion test 결과 출력
print("1 Proportion test")
print('z: {0:0.3f}'.format(stat))
print('p: {0:0.3f}'.format(pval))
```

: 한 집단의 비율이 특정 비율과 같은지를 검정하는 도구가 Proportion test이다. 이때, p-value가 0.05보다 작으므로, 정상이 아닌 동전이다.

## Proportion test (두번째)

```python
# 변수 선언
count = np.array([14,5])
nobs = np.array([1200,1200])

# Proportion test 실행
stat, pval = proportions_ztest(count,nobs)

print("2 Proportion test")
print('z: {0:0.3f}'.format(stat))
print('p: {0:0.3f}'.format(pval))
```

: p-value가 유의수준인 0.05보다 작으므로 두 공장의 불량률이 다르다고 볼 수 있다.
