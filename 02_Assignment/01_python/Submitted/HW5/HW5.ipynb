```python
# 과제1
class Set:
    # parameter가 없을 때는 빈 집합으로 생성
    def __init__(self):
        self.val = list()
        
    # parameter가 있을 때는 중복을 제거하여 생성
    def __init__(self, input_list):
        new_list = []
        for i in input_list:
            if i not in new_list:
                new_list.append(i)
        self.val = new_list
    
    # print
    def __str__(self):
        if len(self.val) == 0:
            return "{}"
        else:
            str_input_set = ""
            for i in self.val[:-1]:
                str_input_set += f"{i}, "
            str_input_set += f"{self.val[-1]}"
            return "{" + str_input_set + "}"
            
    def __len__(self):
        count = 0
        for elem in self.val:
            count += 1
        return count
    
    def __contains__(self, elem):
        if elem in self.val:
            return True
        else:
            return False
    
    # self <= other : self가 other의 부분집합이면 참 반환, 아니면 거짓 반환
    def __le__(self, other):
        for i in self.val:
            if i not in other.val:
                return False
        return True
    
    # other <= self : other이 self의 부분집합이면 참 반환, 아니면 거짓 반환
    def __ge__(self, other):
        for i in other.val:
            if i not in self.val:
                return False
        return True
    
    # a | b
    def __or__(self, other):
        new_list = []
        for i in other.val + self.val:
            if i not in new_list:
                new_list.append(i)
        return Set(new_list)
    
    # a & b
    def __and__(self, other):
        new_list = []
        for i in other.val:
            if i in self.val:
                new_list.append(i)
        return Set(new_list)
    
    # a - b
    def __sub__(self, other):
        new_list = []
        for i in self.val:
            if i not in other.val:
                new_list.append(i)
        return Set(new_list)
    
    # a|= b
    def __ior__(self, other):
        new_list = []
        for i in other.val + self.val:
            if i not in new_list:
                new_list.append(i)
        self.val = new_list
        return self

    # a &= b
    def __iand__(self, other):
        new_list = []
        for i in other.val:
            if i in self.val:
                new_list.append(i)
        self.val = new_list
        return self
    
    # a -= b
    def __isub__(self, other):
        new_list = []
        for i in self.val:
            if i not in other.val:
                new_list.append(i)
        self.val = new_list
        return self
    
    def add(self, elem):
        if elem not in self.val:
            self.val.append(elem)
    
    def discard(self, elem):
        if elem in self.val:
            self.val.remove(elem)
    
    def clear(self):
        self.val = []
```

```python
a = Set([1,2,3,4])
b = Set([1,2,3,4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
