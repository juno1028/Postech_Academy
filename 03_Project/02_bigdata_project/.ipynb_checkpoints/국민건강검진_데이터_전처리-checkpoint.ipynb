### 국민건강검진 결과(표본) - 기본 검진 + 혈액 검사

```python
#데이터 불러오기 
df_h = pd.read_csv("insu_nh_h_screen.csv", engine = "python", encoding="CP949")
```

```python
df_h.info()
```

```python
# 결측 처리 
df_h.isnull().sum(axis=0)
```

```python
df_h.head()
```

```python
# 결측치가 "."로 표시되어 있다. 
```

```python
# 결측치 변환
df_h.replace(".",np.nan,inplace=True)
```

```python
# 검진자 일련 변호 버리기 
del df_h['no']
```

```python
df_h.head()
```

```python
df_h.info()
```

```python
df_h_numeric=df_h.iloc[:,2:21]
```

```python
df_h_numeric.info()
```

```python
df_h_numeric=df_h_numeric.drop(['hearing_left','hearing_right'], axis=1)
```

```python
df_h_numeric.info()
```

```python
df_h_numeric=df_h_numeric.astype('float')
```

```python
df_h_numeric.describe()
```

```python
df_h_numeric.info()
```

```python
df_h_object=df_h.loc[:,['gender','age_group','hearing_left','hearing_right','smoke_flag','drinking_flag']]
```

```python
df_h_object.astype(object)
```

```python
df_h_object.info()
```

```python
# heatmap을 활용한 상관관계 파악 
corr=df_h_numeric.corr()
plt.figure(figsize=(15, 15))
sns.heatmap(corr, vmax=.8, linewidths=0.01,
            square=True,annot=True,cmap='YlGnBu',linecolor="white")
plt.title('Correlation between features');
```

```python
df_result=pd.concat([df_h_numeric,df_h_object])
```
