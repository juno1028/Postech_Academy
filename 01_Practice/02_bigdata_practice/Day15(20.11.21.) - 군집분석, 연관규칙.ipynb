# 군집분석

## 군집분석_연습

```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.cluster import KMeans, DBSCAN
```

```python
df_raw = pd.read_csv("/home/piai/test/Practice/data/직무능력평가.csv", engine="python",encoding="EUC-KR")
df_raw.head()
```

```python
scaler = StandardScaler()
df_raw_scaled = scaler.fit_transform(df_raw)
df_raw_scaled[:5]
```

```python
df_raw_y = df_raw_scaled[:,0]
df_raw_x = df_raw_scaled[:,1::]
```

```python
Z_ward = linkage(df_raw_x, method = "ward", metric = "euclidean")
Z_average = linkage(df_raw_x, method = "average", metric = "euclidean")
Z_minkowski = linkage(df_raw_x, method="average", metric = "minkowski")

plt.figure(figsize=(10,10))
ax = plt.subplot(111)
dendrogram(Z_ward, leaf_font_size = 10, orientation="top")
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("distance")
plt.ylabel("sample index")
```

* 제일 낮게 묶여있는 14,26번부터 묶이기 시작했다는 것.
* 색깔이 두 종류. 두 종류의 군집을 추천한다.

```python
Z_ward
```

```python
cluster_hierarchical = fcluster(Z=Z_ward, t=5, criterion="distance")
df_raw["ClusterHierarchical"] = cluster_hierarchical
df_raw.head()
```

* criterion = "distance" 거리 기반으로, t= 5 : 5개의 군집을 만들었다.

```python
ds_summary_hc = df_raw.groupby("ClusterHierarchical").agg(["min","mean","max"])
level0 = ds_summary_hc.columns.get_level_values(0)
level1 = ds_summary_hc.columns.get_level_values(1)
ds_summary_hc.columns = level0 + "_" + level1

ds_summary_hc.reset_index(inplace = True)
ds_summary_hc.round(3)
```

```python
n_iter_cluster = 15
cluster_range = [i+1 for i in range(n_iter_cluster)]
clus_error = []
for v_n_clus in cluster_range:
    clus = KMeans(v_n_clus)
    clus.fit(df_raw_x)
    clus_error.append(clus.inertia_)
ds_error = pd.DataFrame({"NumberofCluster":cluster_range,"Error":clus_error})
ds_error
```

```python
plt.figure(figsize=(10,10))
plt.plot(ds_error["NumberofCluster"], ds_error["Error"])
plt.title("Sum of squared distance")
plt.xlabel("Clusters")
plt.ylabel("Sum of squared distance")
```

> 군집계수가 증가할수록 Sum of Squared Distance가 줄어든다.

```python
clus = KMeans(5)
clus.fit(df_raw_x)
cluster_kmeans = [i+1 for i in clus.labels_]
df_raw["ClusterKmeans"] = cluster_kmeans
df_raw.head()
```

```python
ds_summary_km = df_raw.drop("ClusterHierarchical", axis=1, inplace=False).groupby("ClusterKmeans").agg(["min","mean","max"])
level0 = ds_summary_km.columns.get_level_values(0)
level1 = ds_summary_km.columns.get_level_values(1)
ds_summary_km.columns = level0 + "_" + level1

ds_summary_km.reset_index(inplace = True)
ds_summary_km.round(3)
```

> 계층형을 먼저 돌려서 대충 군집의 계수를 알아내고, 그 이후에 kmeans의 K를 지정한다.

# 연관규칙 

## 연관규칙_연습

```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df_raw = [["소주","콜라","맥주"],["소주", "콜라", "와인"],["소주","주스"],["콜라","맥주"],["소주","콜라","맥주","와인"],["주스"]]
df_raw
```

```python
enc = TransactionEncoder()
df_raw_enc = enc.fit_transform(X=df_raw)
df_asso = pd.DataFrame(df_raw_enc, columns = enc.columns_)
df_asso.head()
```

```python
v_min_sup = 0.3

df_freq = apriori(df_asso, min_support = v_min_sup, use_colnames = True)
df_freq.round(3)
```

```python
df_asso_rule = association_rules(df_freq, metric="confidence", min_threshold=0.5)
df_asso_rule.round(3)
```

```python
df_asso_rule = association_rules(df_freq, metric="lift", min_threshold=1.5)
df_asso_rule.round(3)
```

:맥주를 사는 사람들에게 콜라도 함께 팔아야한다.

```python
# 고객 ID
ID = list(set(df_raw9["ID"]))
ID.sort()

list_association = []
for i in ID:
    # ID별 구매 데이터 추출
    tmp_list = list(df_raw9[df_raw9["ID"] == i]["PRODUCT"])
    tmp_list.sort()
    # ID별 구매 데이터를 리스트에 담기
    list_association.append(tmp_list)
    
# 리스트 출력
for row in list_association:
    print(row)
```

```python
enc = TransactionEncoder()
df_raw9_enc = enc.fit_transform(X=list_association)
df_asso = pd.DataFrame(df_raw9_enc, columns = enc.columns_)
df_asso.head()
```

```python
v_min_sup = 0.3

df_freq = apriori(df_asso, min_support = v_min_sup, use_colnames = True)
df_freq.round(3)
```

```python
df_asso_rule = association_rules(df_freq, metric="confidence", min_threshold=0.5)
df_asso_rule.round(3)
```

```python
df_asso_rule = association_rules(df_freq, metric="lift", min_threshold=1.5)
df_asso_rule.round(3)
```
